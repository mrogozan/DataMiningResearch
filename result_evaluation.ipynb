```python
import pandas as pd
```

```python
# Convert the time column to datetime if not already
rand_gen_df = pd.read_csv('random_gen_dataset_with_patterns.csv')
rand_gen_df['time'] = pd.to_datetime(rand_gen_df['time'])

# Sort the dataframe by time just to ensure proper order

# Calculate the non-overlapping slice index
rand_gen_df['non_overlapping_slice_index'] = (rand_gen_df.index // 24)
```

```python
# Filter rows where failure == 1
failure_rows = rand_gen_df[rand_gen_df['failure'] == 1]

# Get unique non_overlapping_slice_index values
unique_slice_indices_0 = failure_rows['non_overlapping_slice_index'].unique()

# Print the unique slice indices
print(unique_slice_indices_0)
len(unique_slice_indices_0)
```

```python
def calculate_slices(row_index, chunk_size=24, overlap=0.25):
    slice_list = []
    overlap_size = int(chunk_size * overlap)
    
    for start in range(0, len(rand_gen_df), chunk_size - overlap_size):
        end = start + chunk_size - 1
        if start <= row_index <= end:
            slice_list.append(start // (chunk_size - overlap_size))
    
    return ','.join(map(str, slice_list))

# Apply this function to each row in the dataframe and create the `overlap25_slice` column
rand_gen_df['overlap25_slice'] = rand_gen_df.index.map(calculate_slices)
```

```python
def calculate_slices50(row_index, chunk_size=24, overlap=0.50):
    slice_list = []
    overlap_size = int(chunk_size * overlap)
    
    for start in range(0, len(rand_gen_df), chunk_size - overlap_size):
        end = start + chunk_size - 1
        if start <= row_index <= end:
            slice_list.append(start // (chunk_size - overlap_size))
    
    return ','.join(map(str, slice_list))

# Apply this function to each row in the dataframe and create the `overlap25_slice` column
rand_gen_df['overlap50_slice'] = rand_gen_df.index.map(calculate_slices50)
```

```python
def calculate_slices75(row_index, chunk_size=24, overlap=0.75):
    slice_list = []
    overlap_size = int(chunk_size * overlap)
    
    for start in range(0, len(rand_gen_df), chunk_size - overlap_size):
        end = start + chunk_size - 1
        if start <= row_index <= end:
            slice_list.append(start // (chunk_size - overlap_size))
    
    return ','.join(map(str, slice_list))

# Apply this function to each row in the dataframe and create the `overlap25_slice` column
rand_gen_df['overlap75_slice'] = rand_gen_df.index.map(calculate_slices75)
```
