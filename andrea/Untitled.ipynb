```python
import pandas as pd
```

```python
import pysubgroup as ps
```

```python
df = pd.read_csv('overlap_50_manually_selected_features.csv')
```

```python
df = df.loc[:, (df != df.iloc[0]).any()]
```

```python
df
```

```python
data = df

target = ps.BinaryTarget ('fail_in_slice', 1.0)
searchspace = ps.create_selectors(data, ignore=['fail_in_slice'])
task = ps.SubgroupDiscoveryTask (
    data,
    target,
    searchspace,
    result_set_size=5,
    depth=3,
    qf=ps.SimpleBinomialQF())
results5d5 = ps.GpGrowth().execute(task)
```

```python
pd.options.display.max_colwidth = 500
pd.set_option('display.max_rows', 500)
print(results5d5.to_dataframe())
```
